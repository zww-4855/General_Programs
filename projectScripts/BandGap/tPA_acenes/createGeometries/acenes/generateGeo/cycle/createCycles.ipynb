```python
def output(array,offset,j,outfile):
    of=open(outfile,'a')
    for i in range(len(array)):
        of.write("\n")
        print(array[i][0],array[i][1],array[i][2],array[i][3]+(offset*j))
        of.write(str(array[i][0])+"  "+str(array[i][1])+"  "+str(array[i][2])+"  "+str(array[i][3]+(offset*j)))
```

```python
ucount=2
cellLength=2.415769
output(unitRE,cellLength,0)
for i in range(ucount):
        output(unit,cellLength,i)
output(unitLE,cellLength,ucount)            
```
